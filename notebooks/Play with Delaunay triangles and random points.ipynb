```python
from pyhull.delaunay import DelaunayTri
```

```python
pts = [[-0.5, -0.5], [-0.5, 0.5], [0.5, -0.5], [0.5, 0.5], [0,0]]
```

```python
tri = DelaunayTri(pts)
```

```python
tri.vertices
```

```python
from matplotlib import pyplot as plt

plt.scatter(*zip(*pts))
plt.show()
```

```python
def plot_vertice(a, b, c, points):
    x1, y1 = points[a]
    x2, y2 = points[b]
    x3, y3 = points[c]
    plt.plot([x1, x2, x3, x1], [y1, y2, y3, y1], color="black")
```

```python
for tuple_ in tri.vertices:
    plot_vertice(*tuple_, pts)
```

# From points sampled from curves

```python
import numpy as np

x = 0.1 * np.array(range(1000)) + 1

y = 10 * np.cos(x) ** 2 + np.log(x)

sample_pos = np.random.choice(range(1000), 50)
```

```python
from matplotlib import pyplot as plt

def show_res(x, y, sample_pos=None, name="x"):
    if sample_pos is None:
        sample_pos = list(range(len(x)))
    #plt.scatter(x[sample_pos], y[sample_pos])
    fig = plt.figure()
    pts = list(zip(x[sample_pos], y[sample_pos]))
    tri = DelaunayTri(pts)
    for tuple_ in tri.vertices:
        plot_vertice(*tuple_, pts)
    plt.axis('off')
    fig.savefig(f"plt_{name}.svg")
    
show_res(x, y, sample_pos)
```

```python
x = 10 * np.cos(np.array(range(100))) + 1

y = 10 * np.sin(x) ** 2

sample_pos = list(range(100)) #np.random.choice(range(1000), 100)


show_res(x, y, sample_pos)
```

```python

def tentacle(a=0.1, b=0.8, c=10, d=3, e=30, f=2, size = 300):

    t = a * np.array(range(size)) ** b + 1
    r = c * t ** d + 1
    theta = e * t ** f

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    show_res(x, y, )
    
tentacle()
```

```python
import itertools

a_range = np.linspace(0.01, 0.5, 5)
b_range = np.linspace(0.6, 1.0, 4)
c_range = np.linspace(1, 100, 5)
d_range = np.linspace(2, 4, 5)
e_range = np.linspace(10, 1000, 5)
f_range = np.linspace(1.5, 3, 5)
size_range = np.linspace(300, 700, 5)


iterator = itertools.product(a_range, b_range, c_range, d_range, e_range, f_range, size_range)

```

```python
for i, (a, b, c, d, e, f, size) in enumerate(iterator):
    if i < 10:
        continue
    tentacle(a, b, c, d, e, f, size=size)
    if i > 20:
        break
```
